```python
from sagemaker.estimator import Estimator
```

```python
image_name = '153768889480.dkr.ecr.ap-northeast-2.amazonaws.com/plate-ocr'
role = 'arn:aws:iam::153768889480:role/service-role/AmazonSageMaker-ExecutionRole-20191017T143004'
```

```python
instance_type = 'ml.p2.xlarge'
hpo = {
    "model_dir": '/opt/ml/model',
    "pipeline_config_path": "pipeline.config",
}
```

```python
estimator = Estimator(image_name=image_name,
                      role=role,
                      train_instance_count=1,
                      train_instance_type=instance_type,
                      hyperparameters=hpo
                      )
```

```python
estimator.fit({'train': 's3://7meerkat-dataset/ocr/v0232/train/train.record',
               'val': 's3://7meerkat-dataset/ocr/v0232/val/val.record'})
```
